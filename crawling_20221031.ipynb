```python
import requests
import pandas as pd
```

```python
rank=[]
title=[]
singer=[]
cnt=[]
y=[]

for i in range(2010, 2022):
    year=i
    h= {
        "Referer": "https://circlechart.kr/page_chart/album.circle?ationGbn=T&targetTime={}&hitYear={}&termGbn=year&yearTime=3".format(year,year),
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"
    }
    params={
        "PageSize":"200",
        "curUrl":"circlechart.kr/page_chart/album.circle?ationGbn=T&targetTime={}1&hitYear={}&termGbn=year&yearTime=3".format(year,year),
        "hitYear":year,
        "nationGbn":"T",
        "serviceGbn":"ALL",
        "targetTime":year,
        "termGbn":"year",
        "yearTime":"3"
    }

    url="https://circlechart.kr/data/api/chart/album"
    res=requests.post(url, headers=h ,data=params)
    jsn=res.json()
    li=jsn["List"]
    
    for i in range(0,len(li)):
        rank.append(i+1)
        title.append(li[str(i)]["ALBUM_NAME"])
        singer.append(li[str(i)]["ARTIST_NAME"])
        cnt.append(li[str(i)]["Album_CNT"])
        y.append(year)
len(title)
```

```python
df=pd.DataFrame(rank, columns=['rank'])
df['title']=title
df['singer']=singer
df['cnt']=cnt
df['year']=y
df.to_csv('gaonChart.csv', encoding='utf-8', index = False)
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np



plt.rc("font", family="Malgun Gothic")


while True:
    answer = input("2010 ~ 2021 연도별 TOP 5(A) "
                   "연도별 랭킹 TOP 10(B) "
                   "원하는 가수 앨범 30위 이내 판매량 보기(C) "
                   "원하는 가수 앨범 역대 랭킹 보기(D) "
                   "종료(E) ")

    if answer == "A":
        top5 = []
        
        f = open("gaonChart.csv", encoding="utf-8")
        a = csv.reader(f)
        next(a)  # 헤더빼고 출력
        
        for i in a:
            if int(i[0]) <= 5:
                top5.append(i[1]+" <"+i[2]+">")
        rank = list(range(1, 6))
        year = list(range(2010, 2022))
        df = pd.DataFrame(np.reshape(top5, (12, 5)).T, index=rank, columns=year)
        display(df)
        f.close()

        
    elif answer == "B":
        sell = []
        album = []
        rank = []
        year = input("year(2010 ~ 2021) : ")
        
        f = open("gaonChart.csv", encoding="utf-8")
        a = csv.reader(f)
        next(a)  # 헤더빼고 출력
        
        for i in a:
            if i[4] != "":
                if year in i[4] and int(i[0]) <= 10:
                    sell.append(int(i[3]))
                    rank.append(i[0]+"위")
                    #리스트에 추가
                    album.append(i[1])
        
        for i, v in enumerate(rank):
            plt.text(v, sell[i], album[i], # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
                    fontsize = 9, 
                    horizontalalignment="center",  # horizontalalignment (left, center, right)
                    verticalalignment="bottom")    # verticalalignment (top, center, bottom)

        plt.title(year)
        plt.bar(rank, sell)
        plt.ticklabel_format(style="plain", axis="y")
        plt.show()
        f.close()

        
    elif answer == "C":
        sell = []
        album = []
        singer = input("singer : ")
        
        f = open("gaonChart.csv", encoding="utf-8")
        a = csv.reader(f)
        next(a)  # 헤더빼고 출력
        
        for i in a:  #값을 i에 담음
            if i[4] != "":
                if singer in i[2] and int(i[0]) <= 30:
                    sell.append(int(i[3]))
                    #리스트에 추가
                    album.append(i[1]+"|"+i[4])

        for i, v in enumerate(sell):
            tmp=str(sell[i])+"장"
            plt.text(v, album[i], tmp, # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
                    fontsize = 9, 
                    horizontalalignment="left",  # horizontalalignment (left, center, right)
                    verticalalignment="center")    # verticalalignment (top, center, bottom)

        plt.title(singer)
        plt.barh(album, sell)
        plt.ticklabel_format(style="plain", axis="x")
        plt.show()
        f.close()
        

    elif answer == "D":
        album = []
        rank = []
        singer = input("singer : ")
        
        f = open("gaonChart.csv", encoding="utf-8")
        a = csv.reader(f)
        next(a)  # 헤더빼고 출력
        
        for i in a:  # 값을 i에 담음
            if i[4] != "":
                if singer in i[2]:
                    rank.append(int(i[0]))
                    # 리스트에 추가
                    album.append(i[1] + "|" + i[4])

        for i, v in enumerate(album):
            tmp=str(rank[i])+"위"
            plt.text(v, rank[i], tmp, # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
                    fontsize = 9, 
                    horizontalalignment="center",  # horizontalalignment (left, center, right)
                    verticalalignment="bottom")    # verticalalignment (top, center, bottom)
            
        plt.title(singer)
        plt.plot(album, rank)
        plt.xticks(rotation=90)
        plt.ylim(0,100)
        plt.gca().invert_yaxis()
        plt.show()
        f.close()
        

    elif answer == "E":
        break

    else:
        print("잘못된 입력 값 입니다.")
        continue
```
